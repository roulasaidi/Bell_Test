```python
import numpy as np
import matplotlib.pyplot as plt

# Fonctions  tracer
def e_Chern(delta, N):
    return 0.5 * np.sqrt(-3 * np.log(delta/4) / (0.5 * N))

def e_Cheb(delta, N):
    return np.sqrt(1 / (2 * N * delta))

def e_Hoef1(delta, N):
    return np.sqrt(-np.log(delta/4) / N)

def e_Hoef2(delta, N):
    return  np.sqrt(-np.log(delta/2) *2/ N)

# Paramtres
delta_values = np.linspace(0.01, 0.30, 100)  # Cre 100 valeurs de delta entre 0.01 et 0.30
N = 8000

# Calcul des valeurs pour chaque fonction
e_Chern_values = e_Chern(delta_values, N)
e_Cheb_values = e_Cheb(delta_values, N)
e_Hoef1_values = e_Hoef1(delta_values, N)
e_Hoef2_values = e_Hoef2(delta_values, N)

# Trac des courbes
plt.figure(figsize=(10, 6))
plt.plot(delta_values, e_Chern_values, label='e_Chern')
plt.plot(delta_values, e_Cheb_values, label='e_Cheb')
plt.plot(delta_values, e_Hoef1_values, label='e_Hoef1')
plt.plot(delta_values, e_Hoef2_values, label='e_Hoef2')

# Ajout de titres et de lgendes
plt.title('Values of e as a function of delta')
plt.xlabel('delta')
plt.ylabel('e')
plt.legend()

# Affichage du graphique
plt.show()
```
